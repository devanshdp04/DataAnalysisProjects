# Case Study: Bank Marketing Data Analysis

### Introduction:
This case study involves analyzing and manipulating the "Bank Marketing" dataset. The dataset contains information related to marketing campaigns conducted by a bank to promote term deposits. The objective of this case study is to gain insights into customer behavior, analyze product performance, and understand factors influencing subscription to term deposits.

### Dataset Overview:
The "Bank Marketing" dataset consists of the following columns:

age: Customer's age (numeric)<br>
job: Customer's occupation (categorical)<br>
marital: Marital status of the customer (categorical)<br>
education: Level of education of the customer (categorical)<br>
default: Whether the customer has credit in default or not (categorical)<br>
balance: Customer's account balance (numeric)<br>
housing: Whether the customer has a housing loan or not (categorical)<br>
loan: Whether the customer has a personal loan or not (categorical)<br>
contact: Contact communication type (categorical)<br>
month: Last contact month of the year (categorical)<br>
duration: Duration of the last contact in seconds (numeric)<br>
campaign: Number of contacts performed during this campaign for the customer (numeric)<br>
pdays: Number of days that passed after the customer was last contacted (numeric)<br>
previous: Number of contacts performed before this campaign for the customer (numeric)<br>
poutcome: Outcome of the previous marketing campaign (categorical)<br>
y: Whether the customer subscribed to a term deposit or not (categorical)<br>

## Analyze Customer Behaviour Patterns:

#### Lets do the clustering the data

```python
from sklearn.cluster import KMeans
```

```python
df_x = df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']]
df_x
```

```python
### Apllying Standard scalar method
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X = df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']]
scaled_X = scalar.fit_transform(X)
scaled_X
```

```python
# Applying the K means clustering to the data


kmeans= KMeans(n_clusters= 5)
df['cluster'] = kmeans.fit_predict(scaled_X)
```

```python
# Analyze the distribution of customers across different clusters

sns.countplot(x = 'cluster', data = df)

plt.show()
```

```python
### Analysing the average Age of the customer in each cluster

plt.figure(figsize=(16, 9))
sns.barplot(x = 'cluster', y = 'age', data = df)

plt.title("average Age of the customer in each cluster")
plt.xlabel("Cluster")
plt.ylabel("Average Age")

plt.show()
```
